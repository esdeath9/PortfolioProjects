```python
#created heatmap for better visualisation 

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix for numeric data')


plt.show()
```

```python
#numerised string columns

df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation matrix of all data')


plt.show()
```

```python
correlation_matrix = df_numerized.corr()
corr_pair = correlation_matrix.unstack()

corr_pair
```

```python
corr_sort = corr_pair.sort_values(inplace=False, ascending=False)

with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 8,
                       ):
    print(corr_sort)
```

```python
high_corr = corr_sort[(corr_sort) > 0.5]

high_corr
```
